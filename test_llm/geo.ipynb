```python
import pandas as pd
import json



llama_results = json.load(open("Llama-3.1-8B-Instruct_20251115_041128.json"))
paprika_results = json.load(open("paprika_Meta-Llama-3.1-8B-Instruct_20251115_041253.json"))

llama_df = pd.DataFrame(llama_results["items"])
paprika_df = pd.DataFrame(paprika_results["items"])

llama_df
```

```python
def is_correct(row):
    if row["correct_answer"] == 0 and "a" in row["model_answer_raw"]:
        return True
    elif row["correct_answer"] == 1 and "b" in row["model_answer_raw"]:
        return True
    elif row["correct_answer"] == 2 and "c" in row["model_answer_raw"]:
        return True
    elif row["correct_answer"] == 3 and "d" in row["model_answer_raw"]:
        return True
    elif row["correct_answer"] == 4 and "e" in row["model_answer_raw"]:
        return True
    else:
        return False

llama_df["is_correct"] = llama_df.apply(is_correct, axis=1)
paprika_df["is_correct"] = paprika_df.apply(is_correct, axis=1)
```
